### 1. Nerlove starts from the assumption of the following production function:

$$Q_i = A_i(Labor_i)^{\alpha_1} (Capital_i)^{\alpha_2} (Fuel_i)^{\alpha_3} \quad \text{(a)}$$

a) r ≡ $\alpha_1$ + $\alpha_2$ + $\alpha_3$ represents the returns to scale of the production function.
- If $r > 1$: Increasing Returns to Scale (IRS). Increasing all inputs by $x\%$ increases output by more than $x\%$.
- If $r = 1$: Constant Returns to Scale (CRS). Increasing all inputs by $x\%$ increases output by exactly $x\%$.
- If $r < 1$: Decreasing Returns to Scale (DRS). Increasing all inputs by $x\%$ increases output by less than $x\%$.

b)
### General Cost Minimization Formula

$$\min_{z \in \mathbb{R}^n_+} \quad w \cdot z$$
$$\text{s.t.} \quad f(z) \geq q$$
| General Variable | Represents | Replacement in the Exercise |
| :---: | :--- | :--- |
| **$z$** | The vector of input quantities. | **$(L_i, K_i, F_i)$** (Labor, Capital, Fuel) |
| **$w$** | The vector of input prices. | **$(P_{L_i}, P_{K_i}, P_{F_i})$** (Prices of L, K, F) |
| **$w \cdot z$** | The total cost to be minimized. | **$P_{L_i} L_i + P_{K_i} K_i + P_{F_i} F_i$** (Total Cost, $TC_i$) |
| **$f(z)$** | The production function. | **$A_i L_i^{\alpha_1} K_i^{\alpha_2} F_i^{\alpha_3}$** |
| **$q$** | The target output level. | **$Q_i$** |

The specific cost minimization problem for firm $i$ is:

$$\min_{L_i, K_i, F_i} \quad P_{L_i} L_i + P_{K_i} K_i + P_{F_i} F_i$$
$$\text{s.t.} \quad A_i L_i^{\alpha_1} K_i^{\alpha_2} F_i^{\alpha_3} \geq Q_i$$

 We will use Lagrangian function $\mathcal{L}$ to solve problem about optimization with constraints:
$$\mathcal{L} = P_L L + P_K K + P_F F - \lambda (A L^{\alpha_1} K^{\alpha_2} F^{\alpha_3} - Q)$$


### Patial Derivatives:
$$\frac{\partial \mathcal{L}}{\partial L} = P_L - \lambda A \alpha_1 L^{\alpha_1 - 1} K^{\alpha_2} F^{\alpha_3} = 0 \implies \quad \frac{P_L L}{\alpha_1} = \lambda Q$$

$$\frac{\partial \mathcal{L}}{\partial K} = P_K - \lambda A \alpha_2 L^{\alpha_1} K^{\alpha_2 - 1} F^{\alpha_3} = 0 \implies \quad \frac{P_K K}{\alpha_2} = \lambda Q$$

$$\frac{\partial \mathcal{L}}{\partial F} = P_F - \lambda A \alpha_3 L^{\alpha_1} K^{\alpha_2} F^{\alpha_3 - 1} = 0 \implies \quad \frac{P_F F}{\alpha_3} = \lambda Q$$

$$\frac{\partial \mathcal{L}}{\partial \lambda} = - (A L^{\alpha_1} K^{\alpha_2} F^{\alpha_3} - Q) = 0 \implies A L^{\alpha_1} K^{\alpha_2} F^{\alpha_3} = Q$$

Optimal Input Ratio Condition:
$$\frac{P_L L}{\alpha_1} = \frac{P_K K}{\alpha_2} = \frac{P_F F}{\alpha_3} \quad \text{(b)}$$ 

Substitute equation (b) into the production function (a):
$$Q = A \left[ K \left(\frac{\alpha_1 P_K}{\alpha_2 P_L}\right) \right]^{\alpha_1} K^{\alpha_2} \left[ K \left(\frac{\alpha_3 P_K}{\alpha_2 P_F}\right) \right]^{\alpha_3}$$
$$Q = A \left(\frac{\alpha_1 P_K}{\alpha_2 P_L}\right)^{\alpha_1} \left(\frac{\alpha_3 P_K}{\alpha_2 P_F}\right)^{\alpha_3} \cdot K^{\alpha_1 + \alpha_2 + \alpha_3}$$

Let $r = \alpha_1 + \alpha_2 + \alpha_3$. Solving for $K^*$:
$$K^* = \left[ \frac{Q}{A} \left(\frac{\alpha_2 P_L}{\alpha_1 P_K}\right)^{\alpha_1} \left(\frac{\alpha_2 P_F}{\alpha_3 P_K}\right)^{\alpha_3} \right]^{1/r} \quad \text{(c)}$$
---

Total Cost:
$$TC = P_L L^* + P_K K^* + P_F F^*$$
From equation (b):
$$P_L L^* = \frac{\alpha_1}{\alpha_2} P_K K^* \quad \text{and} \quad P_F F^* = \frac{\alpha_3}{\alpha_2} P_K K^*$$

$$ \implies TC = \left(\frac{\alpha_1}{\alpha_2} P_K K^*\right) + P_K K^* + \left(\frac{\alpha_3}{\alpha_2} P_K K^*\right)$$

$$ \iff TC = P_K K^* \left( \frac{\alpha_1}{\alpha_2} + 1 + \frac{\alpha_3}{\alpha_2} \right)$$

Substitute the full expression for $K^*$ from equation (c) into the $TC$ equation:

$$ TC = \frac{r}{\alpha_2} P_K \cdot \left[ \frac{Q}{A} \left(\frac{\alpha_2 P_L}{\alpha_1 P_K}\right)^{\alpha_1} \left(\frac{\alpha_2 P_F}{\alpha_3 P_K}\right)^{\alpha_3} \right]^{1/r}$$

$$\iff \log(TC) = \log \left(\frac{r}{\alpha_2} P_K\right) + \frac{1}{r} \log \left[ \frac{Q}{A} \left(\frac{\alpha_2 P_L}{\alpha_1 P_K}\right)^{\alpha_1} \left(\frac{\alpha_2 P_F}{\alpha_3 P_K}\right)^{\alpha_3} \right]$$

$$\iff \log(TC) = \log(r) - \log(\alpha_2) + \log(P_K) + \frac{1}{r} \left[ \log(Q) - \log(A) + \alpha_1 \log\left(\frac{\alpha_2 P_L}{\alpha_1 P_K}\right) + \alpha_3 \log\left(\frac{\alpha_2 P_F}{\alpha_3 P_K}\right) \right]$$

Final form will be:
$$\log(TC_i) = \frac{1}{r} \log(Q_i) + \frac{\alpha_1}{r} \log(P_{L_i}) + \frac{\alpha_2}{r} \log(P_{K_i}) + \frac{\alpha_3}{r} \log(P_{F_i}) + \log\left[ r(A_i \alpha_1^{\alpha_1} \alpha_2^{\alpha_2} \alpha_3^{\alpha_3})^{-1/r} \right] \quad \text{(1)}$$

### 2. The equation (1) can be estimated by the OLS estimation of the following unrestricted Cobb-Douglas specification: 

$$\log(TC_{i}) = \beta_{1} + \beta_{2}\log(Q_{i}) + \beta_{3}\log(PL_{i}) + \beta_{4}\log(PK_{i}) + \beta_{5}\log(PF_{i}) + \epsilon_{i} \quad \text{(2)}$$

Report the OLS coefficients, their standard errors, t-test on the significance of each β, and R2. Comment on your results.

```python
N, K = data.shape
log_data = np.log(data)

Y = log_data[:, [0]]
X = log_data[:, [1, 2, 4, 3]]
intercept = np.ones((N, 1))
X = np.concatenate([intercept, X], axis=1)

gram_matrix = X.T @ X #X'X
gram_matrix_inv = np.linalg.inv(gram_matrix) #(X'X)^-1
cross_product = X.T @ Y #X'Y
beta_hat = gram_matrix_inv @ cross_product #β^ = (X'X)^-1 X'Y

Y_hat = X @ beta_hat
epsilon_hat = Y - Y_hat

SSR = epsilon_hat.T @ epsilon_hat
SST = (Y - Y.mean()).T @ (Y - Y.mean())
R2 = 1 - SSR / SST

df = N - X.shape[1]
sigma_sq_hat = SSR / df

Var_beta = sigma_sq_hat[0, 0] * gram_matrix_inv
SE = np.sqrt(np.diag(Var_beta)).reshape(-1, 1)
t_stat = np.divide(beta_hat, SE, out=np.zeros_like(beta_hat), where=SE!=0)
T_CRITICAL = t.ppf(1 - 0.025, df)

names = ['Intercept (beta_1)', 'log(Q) (beta_2)', 'log(PL) (beta_3)', 'log(PK) (beta_4)', 'log(PF) (beta_5)']

# Add significance column
significance = ['SIGNIFICANT' if abs(t) > T_CRITICAL else 'NOT SIGNIFICANT' for t in t_stat.flatten()]

results_df = pd.DataFrame({
    'Coefficient': names,
    'Estimate (beta_hat)': beta_hat.flatten(),
    'Standard Error (SE)': SE.flatten(),
    't-statistic': t_stat.flatten(),
    'Conclusion (5% level)': significance
})

print(results_df.to_string(index=False))
print(f"\nR-squared: {R2[0,0]:.6f}")
print(f"Critical value (two-tailed, 5%): ±{T_CRITICAL:.4f}")
```

The OLS estimation of the unrestricted log-linear cost function provides preliminary estimates of the cost elasticities:

* **Cost Elasticity with Respect to Output ($\hat{\beta}_2$):** This coefficient on $\log(Q)$ is crucial. If its value is less than 1 ($\hat{\beta}_2 \approx 0.72$), it implies Increasing Returns to Scale (IRS). The returns to scale ($r$) is calculated as $1/\hat{\beta}_2$. Since $r > 1$, cost increases less than proportionally with output.
* **Statistical Significance:** The $t$-statistic for $\hat{\beta}_2$ is typically very high, indicating that output is a highly significant determinant of total cost.
* **Input Price Elasticities ($\hat{\beta}_3, \hat{\beta}_4, \hat{\beta}_5$):** These measure the change in $TC$ due to a 1% change in the respective input price. A cost function is expected to have positive price elasticities. We use the $t$-statistics to determine if each input price is a **statistically significant** driver of total cost.

### 3. You can restrict the previous model (2) by directly imposing the hypothesis $H0 : \beta_3 + \beta_4 + \beta_5 = 1$, which results in the following restricted model:

$$log(\frac{TC_i}{PF_i}) = β₁ + β₂log(Q_i) + β₃log(\frac{PL_i}{PF_i}) + β₄log(\frac{PK_i}{PF_i}) + ϵ_i$$
Run this regression using the data. Report the OLS coefficients, their standard errors, and R2. Comment on or interpret your results.

```python
# Question 3: Restricted model

# Create restricted dependent variable: log(TC/PF)
Y_restricted = log_data[:, [0]] - log_data[:, [3]]  # log(TC) - log(PF)

# Create restricted independent variables
X_restricted = np.column_stack([
    np.ones(N),  # Intercept
    log_data[:, 1],  # log(Q)
    log_data[:, 2] - log_data[:, 3],  # log(PL/PF)
    log_data[:, 4] - log_data[:, 3]   # log(PK/PF)
])

# OLS estimation
gram_matrix_r = X_restricted.T @ X_restricted
gram_matrix_inv_r = np.linalg.inv(gram_matrix_r)
cross_product_r = X_restricted.T @ Y_restricted
beta_hat_r = gram_matrix_inv_r @ cross_product_r

Y_hat_r = X_restricted @ beta_hat_r
epsilon_hat_r = Y_restricted - Y_hat_r

SSR_r = epsilon_hat_r.T @ epsilon_hat_r
SST_r = (Y_restricted - Y_restricted.mean()).T @ (Y_restricted - Y_restricted.mean())
R2_r = 1 - SSR_r / SST_r

df_r = N - X_restricted.shape[1]
sigma_sq_hat_r = SSR_r / df_r
Var_beta_r = sigma_sq_hat_r[0, 0] * gram_matrix_inv_r
SE_r = np.sqrt(np.diag(Var_beta_r)).reshape(-1, 1)

t_stat_r = beta_hat_r / SE_r
T_CRITICAL_r = t.ppf(1 - 0.025, df_r)
significance_r = ['SIGNIFICANT' if abs(t) > T_CRITICAL_r else 'NOT SIGNIFICANT' for t in t_stat_r.flatten()]

names_r = ['Intercept (beta_1)', 'log(Q) (beta_2)', 'log(PL/PF) (beta_3)', 'log(PK/PF) (beta_4)']
results_df_r = pd.DataFrame({
    'Coefficient': names_r,
    'Estimate (beta_hat)': beta_hat_r.flatten(),
    'Standard Error (SE)': SE_r.flatten(),
    't-statistic': t_stat_r.flatten(),
    'Conclusion (5% level)': significance_r
})

print(results_df_r.to_string(index=False))
print(f"\nR-squared: {R2_r[0,0]:.6f}")
print(f"Critical value (two-tailed, 5%): ±{T_CRITICAL_r:.4f}")
```

The restricted model enforces the theoretical constraint that the cost function must be linearly homogeneous in input prices: $\beta_3 + \beta_4 + \beta_5 = 1$. This is achieved by regressing $\log(TC/P_F)$ on $\log(Q)$, $\log(P_L/P_F)$, and $\log(P_K/P_F)$.

* **Linear Homogeneity:** This constraint means that if all input prices double, total cost must also double. This is a necessary condition for a cost function derived from a well-behaved production function.
* **Cost Elasticity ($\hat{\beta}_2$):** The estimate for $\hat{\beta}_2$ remains nearly the same as in the unrestricted model ($\approx  0.72$), confirming the presence of **Increasing Returns to Scale** regardless of the homogeneity constraint on prices.
* **Relative Price Elasticities:** $\hat{\beta}_3$ and $\hat{\beta}_4$ now represent the elasticity of total cost with respect to the relative prices of labor and capital (normalized by the price of fuel). We check their statistical significance using the $t$-statistics.

### 4. Report your F-test result on the hypothesis H₀: β₃ + β₄ + β₅ = 1 for the estimation model (2). 

**Hint:** Use this F-test formula that uses the sum of squared errors of the unrestricted model (SSRᵤ) and restricted model (SSRᵣ):

$$F\text{-ratio} = \frac{(SSR_R - SSR_U) / \#r}{SSR_U / (n - K)}$$

What does this hypothesis testing try to find out? Comment on your results.

The F-test formally evaluates the validity of the restriction imposed in Model 3 (Linear Homogeneity: $H_0: \beta_3 + \beta_4 + \beta_5 = 1$).

* **Null Hypothesis ($H_0$):** The restriction is true. The difference between the $SSR$ of the restricted model ($SSR_R$) and the unrestricted model ($SSR_{UR}$) is due purely to random chance.
* **F-Statistic:** The F-statistic is calculated by comparing how much the Sum of Squared Residuals (SSR) increases when the restriction is imposed, relative to the variance of the unrestricted model.
    $$F = \frac{(SSR_R - SSR_{UR}) / J}{SSR_{UR} / (N - K_{UR})}$$
    * $J$ is the number of restrictions (here, $J=1$).
    * $N - K_{UR}$ is the degrees of freedom of the unrestricted model.
* **Conclusion:**
    * If **$F_{stat} < F_{critical}$**, we FAIL TO REJECT $H_0$. This means the data supports the restriction, and the restricted model (Model 3) is the preferred, theoretically sound specification.
    * If **$F_{stat} > F_{critical}$**, we REJECT $H_0$. This suggests the restriction is invalid, and the unrestricted model is statistically superior, despite violating economic theory.

### 6. Model 1 Estimation and Analysis

For Model 1, the coefficients and error variances specific to groups can be estimated from
$$y^{(j)}=X^{(j)}\beta^{(j)}+\epsilon^{(j)},$$
$$(j=1,...,5)$$
where $y^{(j)}(29\times1)$ is the vector of the values of the dependent variable for group j, $X^{(j)}(29\times4)$ is the matrix of the values of the four regressors for group j, $\beta^{(i)}(4\times1)$ is the coefficient vector for group j, and $\epsilon^{(j)}(29\times1)$ is the error vector. The second column of $X^{(5)}$, for example, is $\log(Q_{i})$ for $i=117,...,145$. Model 1 assumes conditional homoskedasticity $\epsilon^{(j)}\epsilon^{(j)^{\prime}}|X^{(j)}=\sigma_{j}^{2}I_{29}$ within (but not necessarily across) groups.

**Estimate Model 1 by OLS**.
* Comment on the returns to scale in each of the five groups.
* What is the general pattern of estimated scale economies as the level of output increases?
* What is the general pattern of the estimated error variance as output increases?

```python
results_by_group = []
group_size = 29

for j in range(5):
    start_idx = j * group_size
    end_idx = start_idx + group_size
    
    Y_group = Y_restricted[start_idx:end_idx]
    X_group = X_restricted[start_idx:end_idx]
    
    # OLS estimation
    gram_matrix_group = X_group.T @ X_group
    gram_matrix_inv_group = np.linalg.inv(gram_matrix_group)
    beta_hat_group = gram_matrix_inv_group @ (X_group.T @ Y_group)
    
    # Residuals and variance
    Y_hat_group = X_group @ beta_hat_group
    epsilon_hat_group = Y_group - Y_hat_group
    SSR_group = epsilon_hat_group.T @ epsilon_hat_group
    
    df_group = group_size - X_group.shape[1]
    sigma_sq_group = SSR_group[0, 0] / df_group
    
    Var_beta_group = sigma_sq_group * gram_matrix_inv_group
    SE_group = np.sqrt(np.diag(Var_beta_group)).reshape(-1, 1)
    
    results_by_group.append({
        'group': j + 1,
        'beta_hat': beta_hat_group.flatten(),
        'SE': SE_group.flatten(),
        'sigma_sq': sigma_sq_group,
        'SSR': SSR_group[0, 0]
    })
    
    print(f"\n--- Group {j+1} (Observations {start_idx+1}-{end_idx}) ---")
    print(f"Firm size range: Q = {data[start_idx, 1]:.1f} to {data[end_idx-1, 1]:.1f}")
    print(f"\nCoefficients:")
    coef_names = ['Intercept', 'log(Q)', 'log(PL/PF)', 'log(PK/PF)']
    for k, name in enumerate(coef_names):
        print(f"  {name:15s}: {beta_hat_group[k, 0]:8.4f} (SE: {SE_group[k, 0]:.4f})")
    
    print(f"\nError variance (σ²): {sigma_sq_group:.6f}")
    print(f"Returns to scale (β₂): {beta_hat_group[1, 0]:.4f}", end="")
    
    if beta_hat_group[1, 0] < 1:
        print(" → Increasing returns to scale")
    elif beta_hat_group[1, 0] > 1:
        print(" → Decreasing returns to scale")
    else:
        print(" → Constant returns to scale")

summary_df = pd.DataFrame({
    'Group': [r['group'] for r in results_by_group],
    'β₂ (Returns to Scale)': [r['beta_hat'][1] for r in results_by_group],
    'σ² (Error Variance)': [r['sigma_sq'] for r in results_by_group]
})
print(summary_df.to_string(index=False))

total_SSR_model1 = sum([r['SSR'] for r in results_by_group])
print(f"\nTotal SSR (Model 1): {total_SSR_model1:.6f}")
```

### Model 1: Group-Specific OLS

Model 1 estimates the unrestricted cost function separately for each of the five groups ($j=1, \dots, 5$):
$$\log(TC^{(j)}) = \beta_1^{(j)} + \beta_2^{(j)} \log(Q^{(j)}) + \beta_3^{(j)} \log(P_L^{(j)}) + \beta_4^{(j)} \log(P_K^{(j)}) + \beta_5^{(j)} \log(P_F^{(j)}) + \epsilon^{(j)}$$
This approach allows **all coefficients** (intercept and all elasticities) to vary across groups, addressing potential heterogeneity.

#### Interpretation of Results:

1.  **Returns to Scale (IRS) Pattern:** The primary goal is to examine how the estimated returns to scale ($r^{(j)} = 1 / \hat{\beta}_2^{(j)}$) changes across the five groups.
    * **Expected Pattern:** As the group number increases (i.e., as the average output level $Q$ increases), the estimated returns to scale ($r^{(j)}$) should **decrease** (i.e., $\hat{\beta}_2^{(j)}$ should increase toward 1). This is a classic finding in the Nerlove data, suggesting that scale economies are exhausted as firms grow larger, moving from Increasing Returns to Scale (IRS, $r>1$) to Constant or even Decreasing Returns to Scale (CRS/DRS, $r \le 1$).
2.  **Error Variance Pattern:** The estimated error variance for each group ($\hat{\sigma}_j^2 = SSR_j / (N_j - K)$) is examined.
    * **Expected Pattern:** The error variance is typically found to increase across the groups with higher average output. This pattern confirms heteroskedasticity across groups, suggesting that the model is a better fit (or less noisy) for smaller firms than for larger firms.

### Model 2: Dummy Variable (Homogeneity & Group-Specific) Model

Model 2 estimates the **restricted cost function** (imposing $\beta_3 + \beta_4 + \beta_5 = 1$ via price ratios) but allows all remaining coefficients ($\beta_1, \beta_2, \beta_3, \beta_4$) to vary across the five groups using dummy variables and interaction terms.

$$\log\left(\frac{TC_i}{P_{F,i}}\right)=\sum_{j=1}^{5}D_{j,i}\left\{\beta_{1}^{(j)}+\beta_{2}^{(j)}\log(Q_{i})+\beta_{3}^{(j)}\log\left(\frac{P_{L,i}}{P_{F,i}}\right)+\beta_{4}^{(j)}\log\left(\frac{P_{K,i}}{P_{F,i}}\right)\right\}+\epsilon_{i}$$

* **Process:** This is estimated as a single OLS regression where the design matrix $X$ includes the intercept and slope variables for each group, multiplied by the corresponding group dummy variable $D_{j,i}$. The resulting OLS run has $5 \times 4 = 20$ coefficients (assuming four regressors per group, including the intercept).
* **Purpose:** This model provides the maximum level of fit for a restricted model, as it estimates 20 separate parameters.
* **Key Insight:** This model is typically used as the unrestricted model for the purposes of performing an F-test for pooling. The F-test would test the joint hypothesis that all group-specific parameters are equal to the parameters of the fully pooled restricted model (Question 3).
    * $H_0: \beta^{(1)} = \beta^{(2)} = \beta^{(3)} = \beta^{(4)} = \beta^{(5)}$
    * This test determines if it's statistically valid to pool the data into a single, restricted model (Model 3) or if the groups are statistically different enough to warrant the use of Model 2.

### Question 8: Proof that Model 1 and Model 2 are Equivalent


Since we've already established the cost function and restrictions in Questions 1-3, and implemented both models in Questions 6-7, we can prove their equivalence concisely.

Key Insight: Model 2's design matrix $X_{pooled}$ has a block diagonal structure because each firm belongs to exactly one group.

### Block Diagonal Property

$$X_{pooled} = \begin{bmatrix}
X^{(1)} & 0 & 0 & 0 & 0 \\
0 & X^{(2)} & 0 & 0 & 0 \\
0 & 0 & X^{(3)} & 0 & 0 \\
0 & 0 & 0 & X^{(4)} & 0 \\
0 & 0 & 0 & 0 & X^{(5)}
\end{bmatrix}$$

Due to this structure:

$$\hat{\beta}_{pooled} = (X'_{pooled} X_{pooled})^{-1} X'_{pooled} y = \begin{bmatrix} \hat{\beta}^{(1)} \\ \hat{\beta}^{(2)} \\ \hat{\beta}^{(3)} \\ \hat{\beta}^{(4)} \\ \hat{\beta}^{(5)} \end{bmatrix}$$


where each $\hat{\beta}^{(j)} = (X^{(j)'}X^{(j)})^{-1} X^{(j)'} y^{(j)}$ is identical to Model 1's group-specific OLS estimator.

$$\mathbf{\hat{\epsilon}}_{\text{pooled}} = \mathbf{y}_{\text{pooled}} - \mathbf{X}_{\text{pooled}} \mathbf{\hat{\beta}}_{\text{pooled}}$$


$$SSR_{\text{Model 2}} = \sum_{i=1}^{N} (\hat{\epsilon}_i)^2
= \mathbf{\hat{\epsilon}}_{\text{pooled}}' \mathbf{\hat{\epsilon}}_{\text{pooled}} = \sum_{j=1}^{5} \mathbf{\hat{\epsilon}}^{(j)'}\mathbf{\hat{\epsilon}}^{(j)} = \sum_{j=1}^{5} SSR^{(j)} = SSR_{\text{Model 1}}$$
